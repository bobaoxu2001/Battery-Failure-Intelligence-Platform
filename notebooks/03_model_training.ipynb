## Train all three models


```python
from src.models.train_soh_model import train as train_soh
from src.models.train_rul_model import train as train_rul
from src.models.train_failure_classifier import train as train_failure
soh = train_soh(); rul = train_rul(); fail = train_failure()
```

## Metrics summary


```python
import pandas as pd
print('SOH  :', soh.algorithm, soh.metrics)
print('RUL  :', rul.algorithm, rul.metrics)
print('FAIL :', fail.algorithm, {k:v for k,v in fail.metrics.items() if k!='confusion_matrix'})
print('Confusion matrix:', fail.metrics['confusion_matrix'])
```

## Feature importance (SHAP if available, else permutation)


```python
fig, axes = plt.subplots(1,3, figsize=(15,4))
for ax, b, name in zip(axes, [soh,rul,fail], ['SOH','RUL','Failure']):
    t = b.importance.head(6)[::-1]
    ax.barh(t.feature, t.importance); ax.set_title(f'{name} drivers ({b.importance.method.iloc[0]})')
plt.tight_layout(); plt.show()
```
